```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Retail Sales Data Analysis/retail_sales.csv')
```

```python
# Remove all null value
df_new = df.dropna()
```

```python
#Finding Total Sales Per Product
Total_sales = df_new.groupby('Product')['Sales'].sum()
print(Total_sales)
```

```python
# Finding The Best selling product
Best_selling = Total_sales.idxmax()
print("Best Selling Product: ", Best_selling)
```

```python
# Computing Average daily sales
daily_sales = df_new['Sales'].mean()
print('Average Daily Sales: ', daily_sales)
```

```python
# Finding Total sales date wise
dly_sales = df_new.groupby('Date')['Sales'].sum()
print(dly_sales)
```

```python
# Ploting Sales Trend Over Time
plt.plot(dly_sales.index, dly_sales.values)
plt.xlabel('Date')
plt.ylabel('Average Sales')
plt.title('Average Sales per Day')
```

```python
# Ploting Total Sales Per product
plt.bar(Total_sales.index, Total_sales.values)
plt.xlabel('Products')
plt.ylabel('Total Sales')
plt.title('Sales per Product')
plt.show()
```
