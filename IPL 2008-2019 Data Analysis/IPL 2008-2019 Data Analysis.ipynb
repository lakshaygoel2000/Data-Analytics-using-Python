```python
# Finding teams preferences of batting or fielding on winning toss
batting_lover = ipl_df[ipl_df['toss_decision']=='bat']
fielding_lover = ipl_df[ipl_df['toss_decision']=='field']
fl = fielding_lover['toss_winner'].value_counts()
bl = batting_lover['toss_winner'].value_counts()
print("Won Toss and choose Fielding \n",  fl,"\n \n Won Toss and choose Batting \n", bl)
```

```python
# Finding common teams in both and sorting alphabetically
common_team = fl.index.intersection(bl.index)
fl = fl[common_team].sort_index()
bl = bl[common_team].sort_index()
fl,bl
```

```python
# Plotting stacked bar chart for common teams for comparing choices  of batting and fielding on winning the toss.
plt.figure(figsize = (17,5), facecolor='lightgray')
plt.bar(list(bl.keys()), list(bl), color = 'khaki', label = 'Bat', )
plt.bar(list(fl.keys()), list(fl), bottom = list(bl), color = 'coral', label = "Field")
plt.xlabel('Teams')
plt.title('Comparision of Batting and fielding')
plt.ylabel('Toss winner choose to Bat vs Field')
plt.legend(["bat", "field"])
plt.xticks(rotation='vertical')
plt.show()
```
